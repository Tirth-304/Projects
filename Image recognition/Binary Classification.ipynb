### Visualise predictions

```python
import numpy as np

def plot_decision_boundary(model , X , y):
    x_min , x_max = X[: , 0].min() - 0.1 , X[: , 0].max() + 0.1
    y_min , y_max = X[: , 1].min() - 0.1 , X[: , 1].max() + 0.1
    xx , yy = np.meshgrid(np.linspace(x_min , x_max , 100),
                          np.linspace(y_min , y_max , 100))
    
    X_in = np.c_[xx.ravel() , yy.ravel()]
    
    y_pred = model.predict(X_in)
    
#     Check for multiclass 
    if(len(y_pred[0]) > 1):
        print('Multilclass')
#         If we are doingmulticlass problem then we might need to change our model\
        y_pred = np.argmax(y_pred , axis = 1).reshape(xx.shape)
     
    else:
        print('Binary class')
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    plt.contourf(xx , yy , y_pred , cmap = plt.cm.RdYlBu , alpha = 0.7)
    plt.scatter(X[: , 0] , X[: , 1] , c = y , s = 40 , cmap = plt.cm.RdYlBu)
    plt.xlim(xx.min() , xx.max())
    plt.ylim(yy.min() , yy.max())
    
```

```python
# Checkout the precision our model is making
# Our model is predicting in linear way while our input is non linear
# So we need to make our model's prediction nonlinear
plot_decision_boundary(model_3 , X , y)
```

* There is a trick to find optimal learning rate that is [learning_rate callback](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/LearningRateScheduler)

### Model - 9 (Find optimum learning rate)

```python
tf.random.set_seed(42)

model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4 , activation = 'relu') ,
    tf.keras.layers.Dense(4 , activation = 'relu') ,
    tf.keras.layers.Dense(1 , activation = 'sigmoid')
])

model_9.compile(loss = 'binary_crossentropy' , 
               optimizer = 'Adam' ,
               metrics = ['accuracy'])

# Traverse a set of learning rate values starting from 1e-4, increasing by 10**(epoch/20) every epoch
lr_schedule = tf.keras.callbacks.LearningRateScheduler(lambda epochs: 1e-4 * 10**(epochs/20))

history = model_9.fit(X_train , y_train , epochs = 100 , callbacks = [lr_schedule])
```

```python
pd.DataFrame(history.history).plot(figsize = (10 , 7))
plt.xlabel('EPOCHS');
```

- As you can see learning rate increases exponentially as number of epochs increases.
- Accuracy also increases(and loss goes down) until specific point when learning rate is increasing slowly.
- To find this point, we can plot loss versus log-scale learning rate.

```python
lrs = 1e-4 * (10 ** (np.arange(100)/20))
plt.figure(figsize = (10 , 7))
# we want the x-axis (learning rate) to be log scale
plt.semilogx(lrs , history.history['loss'])
plt.xlabel('log(Learning rate)')
plt.ylabel('Loss')
plt.title('Learning Rate vs. Loss')
```

To figure out the ideal value of the learning rate (at least the ideal value to begin training our model), the rule of thumb is to take the learning rate value where the loss is still decreasing but not quite flattened out (usually about 10x smaller than the bottom of the curve).

In this case, our ideal learning rate ends up between 0.01 ( 10−2 ) and 0.02.

### Model - 10 (Add Optimal learning rate)

```python
# Let's take 0.02 as ideal learning rate
tf.random.set_seed(42)

model_10 = tf.keras.Sequential([
    tf.keras.layers.Dense(4 , activation = 'relu') ,
    tf.keras.layers.Dense(4 , activation = 'relu') ,
    tf.keras.layers.Dense(1 , activation = 'sigmoid')
])

model_10.compile(loss = 'binary_crossentropy' ,
                optimizer = tf.keras.optimizers.Adam(lr = 0.02) , 
                metrics = ['accuracy'])

model_10.fit(X_train , y_train , epochs = 30)
```
